# 01 — Exploración de señales de voz

En este notebook cargamos los archivos de audio y realizamos una primera inspección visual y auditiva de las señales.

```python
import numpy as np
import matplotlib.pyplot as plt
import librosa
import librosa.display
import soundfile as sf
from pathlib import Path
import IPython.display as ipd

plt.rcParams['figure.figsize'] = (14, 4)
plt.rcParams['figure.dpi'] = 100
```

## 1. Carga de audios

```python
DATA_DIR = Path('../data')
archivos = sorted(DATA_DIR.glob('*.wav'))
print(f'Archivos encontrados: {len(archivos)}')
for f in archivos:
    print(f'  {f.name}')
```

```python
# Cargar el primer archivo como ejemplo
ruta = archivos[0]
y, sr = librosa.load(ruta, sr=None)  # sr=None conserva la tasa original

print(f'Archivo      : {ruta.name}')
print(f'Frecuencia de muestreo : {sr} Hz')
print(f'Duración     : {len(y)/sr:.3f} s')
print(f'Muestras     : {len(y)}')
print(f'Dtype        : {y.dtype}')
```

## 2. Reproducción

```python
ipd.Audio(y, rate=sr)
```

## 3. Visualización en tiempo

```python
t = np.linspace(0, len(y)/sr, num=len(y))

fig, ax = plt.subplots()
ax.plot(t, y, linewidth=0.5, color='steelblue')
ax.set_xlabel('Tiempo [s]')
ax.set_ylabel('Amplitud')
ax.set_title(f'Forma de onda — {ruta.name}')
plt.tight_layout()
plt.savefig('../figuras/01_forma_de_onda.png', dpi=150)
plt.show()
```

## 4. Estadísticas básicas

```python
print(f'Amplitud mínima  : {y.min():.4f}')
print(f'Amplitud máxima  : {y.max():.4f}')
print(f'Media            : {y.mean():.6f}')
print(f'Desv. estándar   : {y.std():.4f}')
print(f'Energía (RMS)    : {np.sqrt(np.mean(y**2)):.4f}')
```

## 5. Histograma de amplitudes

```python
fig, ax = plt.subplots()
ax.hist(y, bins=100, color='steelblue', edgecolor='none', alpha=0.8)
ax.set_xlabel('Amplitud')
ax.set_ylabel('Frecuencia')
ax.set_title('Distribución de amplitudes')
plt.tight_layout()
plt.savefig('../figuras/01_histograma_amplitudes.png', dpi=150)
plt.show()
```
